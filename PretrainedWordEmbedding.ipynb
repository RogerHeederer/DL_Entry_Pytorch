```
from torchtext import data, datasets
```

```
TEXT = data.Field(sequential=True, batch_first=True, lower=True)
LABEL = data.Field(sequential=False, batch_first = True)
```

```
trainset, testset = datasets.IMDB.splits(TEXT, LABEL)
```

```
len(trainset), len(testset)
```

```
trainset[0]
```

```
print(vars(trainset[0]))
```

```
import torch
import torch.nn as nn
from torchtext.vocab import Vectors
```

즉 사전 훈련된 워드투백을 초기 임베딩으로 사용하는 것

```
vectors = Vectors(name = "/content/drive/My Drive/RogerHeederer/DL_entry_PyTorch/eng_w2v")
```

```
#단어 집합을 만듬과 동시에 임베딩 벡터값들 초기화
TEXT.build_vocab(trainset, vectors=vectors, max_size=10000, min_freq=10)
#max_size : 단어 집합 크기 제한
#min_freq : 최소 등장 빈도수 이상만 허용
#vectors = vectors : trainset에서 만들어진 단어 집합의 단어들을. env_w2v에 저장되어 있던 임베딩 벡터값으로 초기화
```

```
TEXT.vocab.stoi['the'],TEXT.vocab.stoi['seeing']
```

```
len(TEXT.vocab) # 10000개의 단어 + <unk> + <pad>
```

```
print('임베딩 벡터의 개수와 차원 : {}'.format(TEXT.vocab.vectors.shape))
```

```
# 임베딩 벡터는 단어의 개수와 마찬가지로 10002개를 가지며, 100차원을 가짐
print(TEXT.vocab.vectors[0]) #<unk>의 임베딩 벡터값
print(TEXT.vocab.vectors[1]) #<pad>의 임베딩 벡터값
print(TEXT.vocab.vectors[296]) # 'seeing' 단어의 임베딩 벡터값
```

```
TEXT.vocab.stoi['self-indulgent']
```

```
#기존에 없었던 단어인 self-indulgent의 임베딩 값 확인해보자, 0으로 초기화 되어 있네.
TEXT.vocab.vectors[10000]
```

```
#위에 만들어진 임베딩 벡터들을 nn.Embedding()의 초기 값으로 사용해보자
embedding_layer = nn.Embedding.from_pretrained(TEXT.vocab.vectors, freeze=False)
```

```
print(embedding_layer(torch.LongTensor([296])))
```
